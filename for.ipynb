## **For 문**

range() 내장 함수는 연속된 숫자를 만드는데 특화되어 있다.  
for문은 반복 실행하는데 특화되어 있다.  
```python 
for 요소를 담을 변수 in 반복가능한 객체:
    반복할 코드(띄어쓰기 4칸)

for 변수 in range(횟수):
    반복할 코드(띄어쓰기 4칸)
```
이 둘을 이용하여 print("Hello world") 1000번 출력하기 구현

````python
for n in range(1000):
    print("Hello world")
````

## range()
range()로 생성된 요소들을 순서대로 하나씩 i변수에 담는다.   
요소를 하나 담을 떄마다 들여쓰기로 구획된 코드 뭉치를 실행 이때 n은 코드뭉치에서 활용 가능 

range는 연속되는 숫자 요소들을 만들 때 활용하면 좋다  
reange()함수는 슬라이싱과 구조가 비슷  
다만, range()의 경우 함수의 형태로 쓰기 때문에 소괄호 안에 구분자는 당연히 ,(콤마)  
시작 인덱스는 포함(이상), 끝 인덱스는 미만  
range()함수를 숫자들의 시퀀스로 iterate할 때 사용하면 편리   
``` python 
range 변수 = range(시작, 끝, 증감크기)
시퀀스자료형변수 [시작:끝:증감크기]
```
연속되는 정수를 생성할 댸 range()내장 함수를 사용하면 편리,  
range()로 반환받은 자료형은 range 이는 list() 내장 함수를 활용해서 리스트로 변경 가능  

아래의 예제는 100미만의 2의 배수를 담은 리스트를 생성하는 방법

````python
even_num = list(range(2, 100, 2))
print(even_num)
````

### range()함수의 특징 
**range() 가 돌려준 객체는 마치 리스트인 것처럼 동작하지만, 사실 리스트와 다르다**

range 객체는 iterate 할 떄 원하는 시퀀스 항목들을 순서대로 요소를 돌려주는 객체지만,  
메모리에 공간을 할당하여 미리 요소들을 만들어 두는 것이 아니라   
필요할 때 그때그때 다음 요소를 만들어 반환하기에 **공간효율적**이다.

java의 일반적인 for문은 for(변수 선언; 조건식; 증감식){실행부}
변수의 숫자를 증감시키고, 산술적 변화에 따라 조건식에 부합할때까지 반복하는 방식

python의 for문은 자바의 향상된 for문과 흡사, 자바의 향상된 for문은 for(자료형 변수명: iterable){실행부}처럼 생겼다. 

- 반복되는 부분을 만들 때 활용
- 혹은 객체(iterable)를 루핑할 때 좋다


#### iterable? 객체?

iterable: 반복 가능한 사전적 의미를 갖는 이 단어는 파이썬에서 이런 뜻을 가진다. 
--> 여러 개의 요소들로 구성되어 있으면서 요소를 한 번에 하나씩 돌려줄 수 있는 객체
예를들어 lsit, str, tupel, range와 같은 시퀀스 자료형이 있다.
